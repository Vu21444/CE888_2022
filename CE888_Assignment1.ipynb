# Combining all CSV files to one dataframe

#### Do Not run the below code for Merging all csv again###

```python
#Path of the csv files
path = r'C:\Users\Vivek\Desktop\Data Science\Assignment\EyeT'

#joining all csv iles
combined_data = pd.DataFrame()

for file in os.listdir(path):
    if file.endswith(".csv"):
        data = pd.read_csv(os.path.join(path, file))
        combined_data = combined_data.append(data)

# save the combined data to a new CSV file
combined_data.to_csv("combined_data.csv", index=False)
```

```python
#load the combined data file
df = pd.read_csv("combined_data.csv")
```

```python
#identifying unique participant name
list=[df["Participant name"].unique()]
list
```

```python
#removing rows that do not have any necessary data
invalid_rows = df[(df['Validity left']=='Invalid') & (df['Validity right']== 'Invalid')]
blank_rows = df[(df['Validity left'] == '') & (df['Validity right'] == '')]
df.replace('', np.nan, inplace=True)
df.drop(invalid_rows.index, inplace=True)
df.drop(blank_rows.index, inplace=True)
```
