```python
import os
import pandas as pd
import numpy as np

# Define the folder path
folder_path = r'C:\Users\Vivek\Desktop\Data Science\Assignment\EyeTdata'

# create backup folder
backup_path = os.path.join(folder_path, 'backup5')
if not os.path.exists(backup_path):
    os.mkdir(backup_path)

# loop through each CSV file in the directory
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)

        # read the CSV file into a dataframe
        df = pd.read_csv(file_path)

        # make a backup copy of the original file
        backup_file = os.path.join(backup_path, filename)
        df.to_csv(backup_file, index=False)

        # replace "," with "." in numeric columns
        numeric_cols = ['Gaze point X', 'Gaze point Y', 'Gaze point left X', 'Gaze point left Y', 'Gaze point right X',
                        'Gaze point right Y', 'Pupil diameter left', 'Pupil diameter right', 'Gaze point X (MCSnorm)',
                        'Gaze point Y (MCSnorm)', 'Gaze point left X (MCSnorm)', 'Gaze point left Y (MCSnorm)',
                        'Gaze point right X (MCSnorm)', 'Gaze point right Y (MCSnorm)']
        df[numeric_cols] = df[numeric_cols].apply(lambda x: x.str.replace(',', '.') if x.dtype == 'object' else x)
        df[numeric_cols] = df[numeric_cols].astype(float)

        # compute mean of numeric columns (excluding null or NaN values) and store in a dictionary
        mean_dict = {}
        for col in numeric_cols:
            mean_dict[col] = df[col].mean(skipna=True)

        # insert mean values at the top of the dataframe
        mean_row = pd.Series(mean_dict, name='Mean')
        df = pd.concat([mean_row.to_frame().T, df], ignore_index=True)

        # remove duplicate values in Project name and Participant name columns
        non_numeric_cols = ['Project name', 'Participant name']
        for col in non_numeric_cols:
            unique_vals = df[col].unique()
            if len(unique_vals) == 1 and not pd.isna(unique_vals[0]):
                df[col] = unique_vals[0]
        
        # save the modified dataframe back to the original file
        df.to_csv(file_path, index=False)
```

```python
# create backup folder
backup_path = os.path.join(folder_path, 'backup')
if not os.path.exists(backup_path):
    os.mkdir(backup_path)

# loop through each CSV file in the directory
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)

        # read the CSV file into a dataframe
        df = pd.read_csv(file_path)

        # make a backup copy of the original file
        backup_file = os.path.join(backup_path, filename)
        df.to_csv(backup_file, index=False)

        # remove 'Validity left' and 'Validity right' columns
        df.drop(['Validity left', 'Validity right'], axis=1, inplace=True)

        # keep only one value for 'Project name' and 'Participant name' columns if all values are the same
        for col in ['Project name', 'Participant name']:
            unique_vals = df[col][1:].unique()
            if len(unique_vals) == 1 and not pd.isna(unique_vals[0]):
                df[col][0] = unique_vals[0]
                df[col][1:] = ''

        # save the modified dataframe back to the original file
        df.to_csv(file_path, index=False)
```

```python
# set the file paths
merged_file_path = 'merged.csv'
questionnaire_file_path = 'Questionnaire_datasetIB.csv'

# create backup folder
backup_path = 'backup9'
if not os.path.exists(backup_path):
    os.mkdir(backup_path)

# make a backup copy of the original files
merged_backup_file = os.path.join(backup_path, 'merged_backup.csv')
questionnaire_backup_file = os.path.join(backup_path, 'questionnaire_backup.csv')
if os.path.exists(merged_file_path):
    os.rename(merged_file_path, merged_backup_file)
if os.path.exists(questionnaire_file_path):
    os.rename(questionnaire_file_path, questionnaire_backup_file)

# read the CSV files into dataframes
merged_df = pd.read_csv(merged_backup_file) if os.path.exists(merged_backup_file) else pd.DataFrame()
questionnaire_df = pd.read_csv(questionnaire_backup_file) if os.path.exists(questionnaire_backup_file) else pd.DataFrame()

# merge the dataframes on the Participant name and Participant nr columns
merged_df['Participant nr'] = merged_df['Participant name'].str.extract(r'(\d+)')
merged_df['Participant nr'] = pd.to_numeric(merged_df['Participant nr'], errors='coerce')
merged_df = merged_df.merge(questionnaire_df[['Participant nr', 'Total Score extended']], on='Participant nr', how='left')

# save the merged dataframe to a CSV file
merged_df.to_csv(merged_file_path, index=False)
```

```python
# load the merged and questionnaire dataframes from the CSV files
merged_df = pd.read_csv('merged_backup.csv', encoding='ascii')
questionnaire_df = pd.read_csv('questionnaire_backup.csv', encoding='windows-1252')

# create a new column in both dataframes with a standardized participant name
merged_df['Participant name'] = merged_df['Participant name'].str.lower().str.replace('participant', '').str.zfill(4)
questionnaire_df['Participant name'] = questionnaire_df['Participant nr'].astype(str).str.zfill(4)

# merge the two dataframes based on the standardized participant name
merged_questionnaire_df = pd.merge(merged_df, questionnaire_df, on='Participant name')

# map the Total Score extended column from questionnaire to the merged dataset based on the Participant nr
merged_questionnaire_df['Total Score extended'] = merged_questionnaire_df['Participant nr'].map(questionnaire_df.set_index('Participant nr')['Total Score extended'])

# save the merged dataframe to a new CSV file
merged_questionnaire_df.to_csv('merged_questionnaire.csv', index=False)
```
